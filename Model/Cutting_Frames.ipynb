```python
import os 
import cv2 

# 이 mp4 파일로 실험 해봄 ( 나중에 사용자 영상이 들어가면 될듯) 
filepath = './Full_Swing.gif' 
video = cv2.VideoCapture(filepath)

# video 가 잘 열렸는지 확인 안되면 종료 
if not video.isOpened() : 
    print("영상이 열리지 않음:", filepath)
    exit() 

# 프레임 저장할 디렉토리 생성 
# filepath[:-4] ->  golf_drivershot.mp4 여기서 .mp4를 뺀거 
#  golf_drivershot 이라는 파일 생성
try : 
    if not os.path.exists(filepath[:-4]): 
        os.makedirs(filepath[:-4])
except OSError : 
    print("Error : 디렉토리오류 "+ filepath[:-4])

# 10프레임(변경가능) 마다 추출 해서 golf_drivershot 파일 에 저장 
# 각 이미지는 frame0.jpg,frame1.jpg, frame2.jpg 이렇게 저장됨
count = 0 
while (video.isOpened()): 
    ret , image = video.read() 
    if( int(video.get(1)) % 1 ==0 ): 
        cv2.imwrite(filepath[:-4] + "/frame%d.jpg" % count , image)
        print("Saved frame number : ", str(int(video.get(1))))
        count += 1 

    # 왜인지 모르겠으나 while 문에서 무한루프 걸려서 
    # 이 조건문으로 종료 
    if video.get(cv2.CAP_PROP_POS_FRAMES) == video.get(cv2.CAP_PROP_FRAME_COUNT):
        break

  
video.release()
```
