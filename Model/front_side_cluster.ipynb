## 결측 이미지 제거 

```python
from PIL import Image
import os

# 이미지가 저장된 디렉토리를 지정합니다.
data_dir = "../UJS/dataset/front"
```

```python
# 디렉토리 내 모든 파일을 순회합니다.
for filename in os.listdir(data_dir):
    # 파일이 이미지인지 확인합니다.
    if filename.endswith(".jpg") or filename.endswith(".png"):
        # 이미지 파일을 엽니다.
        try:
            with Image.open(os.path.join(data_dir, filename)) as img:
                print(f"통과: {filename}은(는) 유효한 이미지 파일입니다.")
        except:
            # 이미지 파일이 유효하지 않으면 오류 메시지를 출력하고 파일을 삭제합니다.
            print(f"오류: {filename}은(는) 유효한 이미지 파일이 아닙니다.")
            os.remove(os.path.join(data_dir, filename))
```

## 데이터셋 생성

```python

# 이미지 크기와 배치 크기 정의
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
```

```python
train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)
try:
    train_generator = train_datagen.flow_from_directory(
        '../UJS/dataset/',
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='binary',
        subset='training') # 훈련용 데이터 생성기
except Exception as e:
    print("Error during training data generation:", e)
    # remove problematic file and try again
    os.remove(e.filename)
    train_generator = train_datagen.flow_from_directory(
        '../UJS/dataset/',
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='binary',
        subset='training')

try:
    validation_generator = train_datagen.flow_from_directory(
        '../UJS/dataset/',
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='binary',
        subset='validation') # 검증용 데이터 생성기
except Exception as e:
    print("Error during validation data generation:", e)
    # remove problematic file and try again
    os.remove(e.filename)
    validation_generator = train_datagen.flow_from_directory(
        '../UJS/dataset/',
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='binary',
        subset='validation')
```

```python

# 모델 정의
model = tf.keras.applications.MobileNetV2(input_shape=(*IMG_SIZE, 3), include_top=False, weights='imagenet')
model.trainable = False
x = tf.keras.layers.GlobalAveragePooling2D()(model.output)
x = tf.keras.layers.Dense(1, activation='sigmoid')(x)
model = tf.keras.models.Model(inputs=model.input, outputs=x)
```

```python

# 모델 컴파일
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# 콜백 함수 정의
class CustomCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        print(f'Epoch {epoch+1}, train loss: {logs["loss"]:.4f}, train accuracy: {logs["accuracy"]:.4f}, val loss: {logs["val_loss"]:.4f}, val accuracy: {logs["val_accuracy"]:.4f}')
```

```python
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python

# 모델 학습
try:
    history = model.fit(train_generator, validation_data=validation_generator, epochs=10)
except PIL.UnidentifiedImageError:
    print('An UnidentifiedImageError occurred during training. Please check the image file format and try again.')
except Exception as e:
    print(f'An error occurred during training: {e}')
```

```python

# 모델 저장
model.save('frist_swing_model.h5')
```
